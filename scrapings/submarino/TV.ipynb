# SUBMARINO - TV

```python
from urllib.request import Request, urlopen
from urllib.error import URLError, HTTPError
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.submarino.com.br/categoria/tv-e-home-theater/tv/l/1p%7CSubmarino/3p%7CCarrefour.com/3p%7CFast%20Shop/3p%7CGAZIN/3p%7CGirafa/g/tipo-de-produto-Hd/tipo-de-produto-Led/tipo-de-produto-Smart%20Tv/tipo-de-produto-Tela/tipo-de-produto-Tela%20Lcd/tipo-de-produto-Tv?ordenacao=relevance&origem=blanca'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}

try:
    req = Request(url, headers = headers)
    response = urlopen(req)
    html = response.read()
    print(html)
    
except HTTPError as e:
    print(e.status, e.reason)
    
except URLError as e:
    print(e.reason) 
```

```python
type(html)
```

```python
html = html.decode('utf-8')
type(html)
```

# Função de Tratamento de Strings

```python
def trata_html(input):
    return " ".join(input.split()).replace('> <','><')
```

```python
html = trata_html(html)
```

# BEAUTIFULSOUP

```python
soup = BeautifulSoup(html, 'html.parser')
```

# WEB SCRAPING

```python
for i in range(10):
    ## obtendo o HTML
    url = 'https://https://www.submarino.com.br/categoria/tv-e-home-theater/tv/l/1p%7CSubmarino/3p%7CCarrefour.com/3p%7CFast%20Shop/3p%7CGAZIN/3p%7CGirafa/g/tipo-de-produto-Hd/tipo-de-produto-Led/tipo-de-produto-Smart%20Tv/tipo-de-produto-Tela/tipo-de-produto-Tela%20Lcd/tipo-de-produto-Tv/pagina-'+str(i+1)+'?ordenacao=relevance&origem=blanca'
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}

    try:
        req = Request(url, headers = headers)
        response = urlopen(req)
        html = response.read()
        print(html)
    
    except HTTPError as e:
        print(e.status, e.reason)

    except URLError as e:
        print(e.reason)
        
    for anuncio in anuncios:
        card = {}

        card['Descrição'] = anuncio.find('h2', {'class': 'TitleUI-bwhjk3-15 jfuFNN TitleH2-sc-1wh9e1x-1 fINzxm'}).getText()


        card['Preço'] = anuncio.find('div', {'class': 'PriceWrapper-bwhjk3-13 eBwWGp ViewUI-sc-1ijittn-6 iXIDWU'}).getText()

        try:
            card['TotalReviews'] = anuncio.find('span', {'class': 'Quantity-sc-1fg2071-3 RsSwi TextUI-sc-12tokcy-0 gZkeKo'}).getText().replace('(','').replace(')','')

        except AttributeError:
            card['TotalReviews'] = 0


        cards.append(card)
```

```python
dataset_TV = pd.DataFrame(cards)
dataset_TV
```
