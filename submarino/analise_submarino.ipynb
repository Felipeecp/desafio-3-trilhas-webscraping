```python
import pandas as pd
```

```python
dados_smartphone = pd.read_csv('output\data\dataset_smartphone.csv',sep=';')
dados_notebook = pd.read_csv('output\data\dataset.csv',sep=';')
dados_tv = pd.read_csv('output\data\dataset_TV.csv',sep=';')
dados_geladeira = pd.read_csv('output\data\dataset_Geladeira.csv',sep=';')
```

```python
dados_submarino = pd.concat([dados_notebook,dados_smartphone,dados_tv,dados_geladeira])
dados_submarino.sample(5)
```

```python
dados_submarino.shape
```

```python
dados_submarino['Preço'] = dados_submarino['Preço'].apply(lambda x: x.replace('R$ ','')).apply(lambda x: x.replace('.','')).apply(lambda x: x.replace(',','.'))
dados_submarino['Preço']=dados_submarino['Preço'].apply(pd.to_numeric, errors='coerce')
```

```python
dados_submarino
```

```python
dados_submarino.to_csv('output\data\dados_submarino.csv',sep=';',index=False)
```
