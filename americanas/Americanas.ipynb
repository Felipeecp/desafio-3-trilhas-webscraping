```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
def tratar_html(dados)->str:
    """Função para tirar os \n e \t"""
    return " ".join(dados.split()).replace("> <", "><")

def pegarHtml(url)->BeautifulSoup:
    driver = webdriver.Chrome()
    driver.get(url)
    #driver.implicitly_wait(10)
    time.sleep(10)
    robo_html = driver.find_element_by_css_selector("div.grid__StyledGrid-sc-1man2hx-0.iFeuoP")
    html = robo_html.get_attribute("outerHTML")
    return BeautifulSoup(html, "html.parser")
```

```python
Listacards = []
```

```python
url = "https://www.americanas.com.br/busca/smartphone?limit=24&offset=0"
soup = pegarHtml(url)
```

```python
cards = soup.find_all("div", {"class": "src__Wrapper-sc-1l8mow4-0"})
```

```python
for card in cards:
    url_page = "https://www.americanas.com.br/" + card.find("a").get("href")
    img = card.find("img").get("src")
    name = card.find("h3", {"class": "product-name__Name-sc-1shovj0-0"}).getText()
    avaliacao = card.find("span", {"class": "src__Count-r5o9d7-1"})
    avaliacao = avaliacao.getText().split()[0]
    try:
        avaliacao = int(avaliacao)
    except:
        avaliacao = 0
        
    preco_info = card.find("div", {"class": "price-info__Wrapper-sc-1xm1xzb-0"})
    preco_real = preco_info.find("span", {"class": "price__Price-h6xgft-0"})
    
    if preco_real != None:
        preco_real = preco_real.text.split()[1]
    preco_promotion = preco_info.find("span", {"class": "price__PromotionalPrice-h6xgft-1"})
    if preco_real != None:
        preco_real = preco_promotion.text.split()[1]
    
    #Agrupando dados
    cardDetail = {'url_page': url_page, "img": img, "name": name, "avaliacao": avaliacao,"preco_real":  preco_real, "preco_promotion": preco_promotion}
    
    Listacards.append(cardDetail)
```
